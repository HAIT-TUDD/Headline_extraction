```python
import requests
import csv
import shutil
import time
from tqdm import tqdm
```

```python
metadata= {}
with open("SK3_Artikel.txt", encoding = "utf-8", newline= "") as csvdatei:
    csv_reader_object = csv.reader(csvdatei, delimiter = ";")
    next(csv_reader_object, None)
    for row in csv_reader_object:
        date = row[2].split(" ")[0]
        kat = row[6]
        article_title = row[1]
        article_page = row[5].split(",")[0]
        article_id = row[3]

        if date in metadata:

            if article_id in metadata[date]:
                metadata[date][article_id][0].append(kat)

            else:
                metadata[date][article_id] = [[kat], article_title, article_page]
        
        else:
             metadata[date] = {article_id: [[kat], article_title, article_page] }
```

```python
metadata["13031936"]
```

```python
#https://hait.tu-dresden.de/slub/1936/10/files/06/files/00000005.tif.large.jpg

url_list =[]
headers = {}
url = "https://hait.tu-dresden.de/slub/1936/"
for year in metadata.keys():
    if "1936" in year:
        titlelist = []
        for k,v in metadata[year].items():
            month = year[2:4]
            day = year[0:2]
            page = v[2] if len(v[2]) == 2 else "0" + v[2]
            title = v[1]
            titlelist.append([title,page])
            url_list.append(url + month + "/files/" + day + "/files/" + "000000" + page + ".tif.large.jpg")
        
        headers[year] = titlelist
            
```

```python
headers
```

```python
with open('titles.csv', 'w', encoding='utf-8') as f:
    w = csv.writer(f)
    w.writerow(["id", "title", "page"])
    for k,v in headers.items():
        for t in v:
            w.writerow([k,t])
```

```python
len(url_list)
```

```python

for url in tqdm(url_list):
    res = requests.get(url, stream = True)
    print(res)
    
    file_name = url.split("/")[7] + "_" + url.split("/")[5] + "_" + url.split("/")[4] + "_" + url.split("/")[9]

    
    if res.status_code == 200:
        with open("./Pages/" + file_name,'wb') as f:
            shutil.copyfileobj(res.raw, f)
        print('Image sucessfully Downloaded: ', "./Pages/" + file_name)
    else:
        print('Image Couldn\'t be retrieved')
        
    time.sleep(0.5)
```
